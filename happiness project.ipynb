```python
#finding best random_state
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr=LinearRegression()
maxacc=0
maxrs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    acc=r2_score(y_test,y_pred)
    if acc > maxacc:
        maxacc=acc
        maxrs=i
print("best accuracy is",maxacc,"on random_state",maxrs)
        
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=37)
```

```python
#we are not doing scaling due to all the data are in same range
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso
ls=Lasso(alpha=0.001)
ls.fit(x_train,y_train)
y_pred=ls.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
rg=Ridge(alpha=0.001)
rg.fit(x_train,y_train)
y_pred=rg.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import ElasticNet
es=ElasticNet(alpha=0.001)
es.fit(x_train,y_train)
y_pred=es.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostRegressor
ad=AdaBoostRegressor(random_state=37,n_estimators=100)
ad.fit(x_train,y_train)
y_pred=ad.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gd=GradientBoostingRegressor(learning_rate=0.1,n_estimators=100)
gd.fit(x_train,y_train)
y_pred=gd.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#minimum difference is the accuracy,so ElasticNet is the best model, we use for this dataset.so we have to make it hyperparameter_tuning
```

```python
from sklearn.model_selection import GridSearchCV
ratio=np.arange(0,1,0.1)
alpha=[1e-5,1e-4,1e-4,1e-3,1e-1,0,0.1,1.0,100]
parameter={'l1_ratio':ratio,'alpha':alpha}
gscv=GridSearchCV(ElasticNet(),parameter,cv=5)
gscv.fit(x_train,y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
gscv.best_estimator_
```

```python
gscv_pred=gscv.best_estimator_.predict(x_test)
```

```python
r2_score(y_test,gscv_pred)
```
